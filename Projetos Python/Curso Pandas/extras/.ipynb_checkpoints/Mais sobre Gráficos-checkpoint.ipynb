```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor x Área')

g2.hist(dados.Valor)
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra Valor')

grupo = dados.groupby('Tipo')['Valor']
labels = grupo.mean().index
values = grupo.mean().values
g4.bar(labels, values)
g4.set_title('Média por Tipo')
```

```python
area.savefig('grafico.png', dpi = 300, bbox_inches = 'tight')
```

```python
dados = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)

dados_g1 = dados.groupby('Tipo Agregado')['Valor']
labels = dados_g1.count().index
values = dados_g1.count().values
g1.pie(values, labels = labels, autopct='%1.1f%%')
g1.set_title('Total de Imóveis por Tipo Agregado')

dados_g2 = dados.groupby('Tipo')['Valor']
labels = dados_g2.count().index
values = dados_g2.count().values
g2.pie(values, labels = labels, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Imóveis por Tipo')
```
